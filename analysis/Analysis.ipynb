<h4> Label Encoding </h4>

```python
from sklearn import preprocessing
label_encoder_features = preprocessing.LabelEncoder()
label_encoder_target = preprocessing.LabelEncoder()

# encode module names
label_encoder_features.fit(modules)
label_encoder_target.fit(grades)

df_1['Module'] = label_encoder_features.transform(df_1['Module'])
df_1['Module'] = df_1['Module'].astype("category")

df_1['Final Grade'] = label_encoder_target.transform(df_1['Final Grade'])
df_1['Final Grade'] = df_1['Final Grade'].astype("category")

df_1.head(5)
```

<h4> Split dataset for traning and testing </h4>

```python
from sklearn.model_selection import train_test_split

# define y and X variables
X = df_1.drop(['Final Grade','End Marks', 'Final Marks'], 1)
y = pd.DataFrame(df_1['Final Grade'], columns=['Final Grade'])

# spliting dataset
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 101)

# reset indexes
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

X_train.head(5)
```

```python
print(f'Train Dataset (X) = {X_train.shape}, (y) = {y_train.shape}')
print(f'Test Dataset (X) = {X_test.shape}, (y) = {y_test.shape}')
```

<h4> Feature Scaling </h4>

```python
from sklearn.preprocessing import MinMaxScaler

# Removing Categorical Features before the feature scaling
columns = X_train.columns
# Continuous col
columns_categorical = np.delete(columns,np.s_[1:])
# Categorical col
columns_cont = np.delete(columns,np.s_[0])

# Applying Standardization 
# Init MinMaxScaler
scaler_features = MinMaxScaler() 

# Fit and Transform training dataset features
X_train_except = pd.DataFrame(X_train, columns = columns_cont)
scaler_features.fit(X_train_except)
X_train = pd.DataFrame(scaler_features.transform(X_train_except), columns = columns_cont).join(X_train[columns_categorical])

# Transform testing dataset features
X_test_except = pd.DataFrame(X_test, columns = columns_cont)
X_test = pd.DataFrame(scaler_features.transform(X_test_except), columns = columns_cont).join(X_train[columns_categorical])

X_train.head(5)
```

<h2> Decision Tree Classifier </h2>

```python
# build model

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=123)
model.fit(X_train,y_train)
predictions = model.predict(X_test)

predictions
```

```python
predictions_labels = label_encoder_target.inverse_transform(predictions)
actual_labels = label_encoder_target.inverse_transform(y_test)


TP = 0
FP = 0
for i in range(len(predictions_labels)):
    print(f'Predicted = {predictions_labels[i]}, Actual = {actual_labels[i]}')
    if predictions_labels[i] == actual_labels[i]:
        TP += 1
    else:
        FP += 1
```

```python
'''
    Function for determine grade
    @param
        score - int or float number between 0 - 100
'''
def grade(score):
    if score < 0.0:
        raise Exception("Score range must be in 0-100")
    elif score < 35.0:
        grade = 'D'
    elif score < 40.0:
        grade = 'C-'
    elif score < 45.0:
        grade = 'C'
    elif score < 50.0:
        grade = 'C+'
    elif score < 55.0:
        grade = 'B-'
    elif score < 60.0:
        grade = 'B'
    elif score < 65.0:
        grade = 'B+'
    elif score < 75.0:
        grade = 'A-'
    elif score < 85.0:
        grade = 'A'
    elif score <= 100.0:
        grade = 'A+'
    else:
        raise Exception("Score range must be in 0-100")
    return grade
```

```python
TP = 0
FP = 0
for i in range(len(predictions)):
    p_value = predictions[i]
    a_value = actual[i]
    
    assignment = X_test_actual[i][0]
    quiz = X_test_actual[i][1]
    midTerm = X_test_actual[i][2]
    miniProject = X_test_actual[i][3]
    
    score = (assignment + quiz) * 0.1 + midTerm * 0.1 + miniProject * 0.2
    
    predicted_score = score + 0.6 * p_value
    actual_score = score + 0.6 * a_value
    
    p_grade = grade(predicted_score)
    a_grade = grade(actual_score)
    
    print(f'Actual = {a_grade}, Predicted = {p_grade}')
    
    if p_grade == a_grade:
        TP += 1
    else:
        FP += 1
```
